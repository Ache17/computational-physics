### Problem 3-4 

```python
import plotly.graph_objects as go
import numpy as np

x = np.linspace(-10, 10, 5000)

def psi1(x):
    return np.exp(-x**4)

def psi2(x):
    return np.exp(-x**6)

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=psi1(x), name="$ e^{-x^{4}} $"))
fig.add_trace(go.Scatter(x=x, y=psi2(x), name='$e^{-x^{6}}$'))
fig.show()
```

Normalize wavefunction 
$$ \int_{-\infty }^{\infty } \Psi \Psi^{*} dx  = 1$$

```python
from scipy.integrate import quad

psi1_squared = lambda x: psi1(x)**2
psi2_squared = lambda x: psi2(x)**2

N1 = quad(psi1_squared, -np.inf, np.inf)[0]
N2 = quad(psi2_squared, -np.inf, np.inf)[0]

print("N^{2}" + " for psi 1 = {}".format( N1 ))
print("N^{2}" + " for psi 2 = {}".format( N2 ))
```

```python

psi1_squared_normalized = lambda x: psi1_squared(x) / N1
psi2_squared_normalized = lambda x: psi2_squared(x) / N2

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=psi1_squared_normalized(x), name="$ e^{-x^{4}} \cdot e^{-x^{4}} $"))
fig.add_trace(go.Scatter(x=x, y=psi2_squared_normalized(x), name='$e^{-x^{6}} \cdot e^{-x^{6}}$'))
fig.show()
```

Calculating k-space distribution of given wavefunction is calculated from 
$$ \Phi(k) = \frac{1}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} \psi(x,0) e^{-ikx} dx $$

```python
phi1 = lambda x: psi1(x) * np.exp(-1j * x * k)
phi2 = lambda x: psi2(x) * np.exp(-1j * x * k)

phi1_k = []
phi2_k = []
ks = np.linspace(-100, 100, 1000)
for k in ks:
    phi1_k.append(quad(phi1, -np.inf, np.inf)[0])
    phi2_k.append(quad(phi2, -np.inf, np.inf)[0])

fig = go.Figure()
fig.add_trace(go.Scatter(x=ks, y=phi1_k / np.sqrt(2 * np.pi), name="$\phi_{1}(k)$"))
fig.add_trace(go.Scatter(x=ks, y=phi2_k / np.sqrt(2 * np.pi), name="$\phi_{2}(k)$"))
fig.show()
```

### Problem 10

```python
a = 1
xp = 0

def psi(x, tau):
    xsip = xp/a
    xsi = x/a
    
    integral_fn = lambda val: np.exp(np.sqrt(3) * 1j * (xsi - val)**2 / tau )
    integral = quad(integral_fn, -0.5, 0.5, limit=1000)
    mult = np.sqrt(np.sqrt(3) / (np.pi * a * tau * 1j))
    val = mult * integral[0]
    return (val * np.conjugate(val)).real

fig = go.Figure()
taus = [0.001,0.01,0.1, 0.25, 0.5, 1]
for tau in taus:
    xs = np.linspace(-3,3, 500)
    ys = []
    for x in xs:
        ys.append(psi(x, tau))
    fig.add_trace(go.Scatter(x=xs, y=ys, name="tau="+ str(tau)))
fig.show()
```

```python
taus_pl = []
taus = np.linspace(0, 0.5, 5000)

fig = go.Figure()

for tau in taus:
    taus_pl.append(psi(0, tau))

fig.add_trace(go.Scatter(x=taus, y=taus_pl))
fig.show()
```
